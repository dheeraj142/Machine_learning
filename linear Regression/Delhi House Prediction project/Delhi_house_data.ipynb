```python
def ret_1(x):
    l=x.split(',')
    return l[0]
```

```python
df3 = df2.copy()
```

```python
df3.head(2)
```

```python
df3.Address = df3.Address.apply(ret_1)
```

```python
df3.Address = df3.Address.apply(lambda x: x.strip())
```

```python
add_stats = df3.groupby('Address')['Address'].agg('count').sort_values(ascending=False)
```

```python
add_stats
```

```python
len(add_stats[add_stats<=10])
```

```python
loc_less_than_10 = add_stats[add_stats<=10]
```

```python
loc_less_than_10
```

```python
df3.Address = df3.Address.apply(lambda x: 'other' if x in loc_less_than_10 else x)
```

```python
df3
```

```python
df4 = df3.copy()
```

```python
df4['total_sqft'] = df3['price']/df3['Price_sqft']
```

```python
df4.head()
```

```python
df4.shape
```

```python
df4[df4.total_sqft/df4.Bedrooms<300].head()
```

```python
df5 = df4[~(df4.total_sqft/df4.Bedrooms<300)]
```

```python
df5.shape
```

```python
df5.Price_sqft.describe()
```

```python
def rem_out(df):
    df_out = pd.DataFrame()
    for key,subdf in df.groupby('Address'):
        m = np.mean(subdf.Price_sqft)
        std = np.std(subdf.Price_sqft)
        reduced_df = subdf[(subdf.Price_sqft>(m-std)) & (subdf.Price_sqft<=(m+std))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
```

```python
df6 = rem_out(df5)
```

```python
df6
```

```python
df6.Bedrooms =df6.Bedrooms.apply(lambda x: int(x))
```

```python
df6.head()
```

```python
df6.Bathrooms = df6.Bathrooms.apply(lambda x: int(x))
```

```python
df6.head(2)
```

```python
def plot(df,loc):
    bd2 = df[(df.Address==loc)&(df.Bedrooms==2)]
    bd3 = df[(df.Address==loc)&(df.Bedrooms==3)]
    matplotlib.rcParams['figure.figsize']=(5,10)
    plt.scatter(bd2.total_sqft,bd2.price,color='blue',label='2 Bedrooms',s=50)
    plt.scatter(bd3.total_sqft,bd3.price,color='green',marker="+",label='3 Bedrooms',s=50)
    plt.xlabel("Total sqft")
    plt.ylabel("Price")
    plt.title(loc)
    plt.legend()
```

```python
plot(df6,'Raj Nagar Extension')
```

```python
plot(df6,'Indirapuram')
```
